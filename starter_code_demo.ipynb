```python
from cyl_heat import construct_system
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_r = 25
N_z = 25

r_edge = np.linspace(0, 1, N_r+1)
z_edge = np.linspace(0, 1, N_z+1)

r_center = r_edge[1:]-r_edge[0]/2
z_center = z_edge[1:]-z_edge[0]/2

```

```python
kfun = lambda r, z : 1.23
qppp = lambda r, z : 0.0
```

```python
BC_z = ('dirichlet', 0)
BC_r_top = ('dirichlet', 0)
BC_r_bottom = ('dirichlet', 1)
```

```python
A, b = construct_system(r_center, z_center, kfun, qppp,
                        BC_z, BC_r_top, BC_r_bottom)
```

```python
plt.spy(A)
```

```python
T = np.linalg.solve(A, b)
```

```python
plt.imshow(T.reshape((N_z, N_r)), cmap="inferno")
```
